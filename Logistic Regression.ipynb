```python
def Age (Col):
    if (pd.isnull(Col[1])):
        if (Col[0] == 1 ): 
            return 37
        elif (Col[0] == 2):
            return 29
        else:
            return 27
    else: 
        print(Col)
        return Col[1]
 
        

train[['Pclass','Age']].apply(Age,axis=1)
```

```python
train['Age']
```

```python
sns.heatmap(train.isnull(),cmap='viridis')
```

```python
train.drop('Cabin',axis=1,inplace = True)
```

```python
train.dropna(inplace = True)
```

```python
sex = pd.get_dummies(train['Sex'],drop_first=True)
e = pd.get_dummies(train['Embarked'])
train = pd.concat([train,sex,e],axis=1)
train.drop('C',axis=1)
```

```python
train['Sex']=pd.get_dummies(train['Sex'],drop_first=True)
```

```python
train[['Q','S']]=pd.get_dummies(train['Embarked'],drop_first=True)
```

```python
train.drop(['PassengerId','Name','Ticket','Embarked'],axis=1,inplace=True)
```

```python
train
```

```python
X = train.drop(['Survived'],axis=1)
```

```python
y = train['Survived']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
prediction = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(prediction,y_test))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(prediction,y_test)
```
